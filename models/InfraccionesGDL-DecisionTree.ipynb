##### 6. Realizar varias veces los siguientes pasos, para tratar de encontrar el mejor modelo de predicción:

###### a) Prueben todos los modelos de clasificación vistos en clase para generar predicciones. Al utilizar árboles de decisión, genera también la imagen del mismo.

```python
clean_training = clean_training.dropna(subset = 'LATITUD')

x_alfredo = clean_training.drop(['INFRACCION'], axis=1)

str_features = ['ESTADO', 'MARCA']

for i in range(len(str_features)):
    clean_training[str_features[i]] = clean_training[str_features[i]].astype("category")
    clean_training[str_features[i]+"_cat"] = clean_training[str_features[i]].cat.reorder_categories(clean_training[str_features[i]].unique(),
                                                                       ordered=True)
    clean_training[str_features[i]+"_cat"] = clean_training[str_features[i]+"_cat"].cat.codes

x_alfredo['MES'] = x_alfredo.FECHA.dt.month

for i in range(len(str_features)):
    x_alfredo[str_features[i]] = x_alfredo[str_features[i]].astype("category")
    x_alfredo[str_features[i]+"_cat"] = x_alfredo[str_features[i]].cat.reorder_categories(x_alfredo[str_features[i]].unique(),
                                                                       ordered=True)
    x_alfredo[str_features[i]+"_cat"] = x_alfredo[str_features[i]+"_cat"].cat.codes

x_alfredo['MES'] = x_alfredo.FECHA.dt.month

x_alfredo_all = x_alfredo
```

1. Decision Tree

```python
x_alfredo = x_alfredo[['ESTADO', 'MARCA', 'CALLE', 'CRUCE', 'FECHA', 'ANO', 'MES',
                 'HORA', 'HORA_NUM', 'LATITUD', 'LONGITUD']]
y_alfredo = clean_training['INFRACCION']
y_alfredo = y_alfredo.replace({1: 0, 4: 1, 5: 2})

X_train_tree_alfredo, X_test_tree_alfredo, y_train_tree_alfredo, y_test_tree_alfredo = train_test_split(x_alfredo, y_alfredo, test_size=0.25, random_state=1)
```

```python
X_train_tree_alfredo.head()
```

```python
import category_encoders as ce

encoder_gnb_alfredo = ce.OrdinalEncoder(cols=['ESTADO', 'MARCA', 'CALLE', 'CRUCE', 'FECHA',
                                           'ANO', 'MES', 'HORA', 'HORA_NUM', 'LATITUD', 'LONGITUD'])
```

```python
X_train_tree_alfredo = encoder_gnb_alfredo.fit_transform(X_train_tree_alfredo)
X_test_tree_alfredo = encoder_gnb_alfredo.transform(X_test_tree_alfredo)
```

```python
X_train_tree_alfredo.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier(max_depth=6, random_state=42)
tree_model.fit(X_train_tree_alfredo, y_train_tree_alfredo)
```

```python
# predicciones
y_pred_alfredo = tree_model.predict(X_test_tree_alfredo)
```

###### b) Calculen y utilicen métricas para comparar los modelos: accuracy, matriz de confusión, precisión, sensitividad, especificidad.

1. Decision Tree

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

print('Classification report')
print(classification_report(y_test_tree_alfredo, y_pred_alfredo))

print('R2 Score')
print(r2_score(y_test_tree_alfredo, y_pred_alfredo), '\n')

print('Mean Squared Error')
print(np.sqrt(mean_squared_error(y_test_tree_alfredo, y_pred_alfredo)), '\n')

cm_alfredo = confusion_matrix(y_test_tree_alfredo, y_pred_alfredo)
print('Confusion matrix\n\n', cm_alfredo)
```
